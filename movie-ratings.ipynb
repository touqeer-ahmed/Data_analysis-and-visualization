```python
list1 = []
genre = []
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre==gen].Budget)
    genre.append(gen)
sns.set_style('whitegrid')
fig, ax = plt.subplots()
fig.set_size_inches(12,8)
h=plt.hist(list1,bins= 30,stacked=True,rwidth=1,label= genre)
plt.legend(prop = {'size':20},shadow=True)
plt.title('Movie Budget Distribution',fontsize= 30,color = 'DarkBlue')
plt.xlabel('Budget',fontsize=15,color = 'Green')
plt.ylabel('No of movies',fontsize =15,color = 'Red')
plt.xlim(0,330)
plt.show()
```
